```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/CO2 Emissions_Canada.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns="Transmission")
```

```python
bool_cols = df.select_dtypes(include='bool').columns
print(bool_cols)
```

```python
df.info()
```

```python
df['Make'] = df['Make'].astype(str)
df['Model'] = df['Model'].astype(str)
df['Vehicle Class'] = df['Vehicle Class'].astype(str)
```

```python
df = pd.get_dummies(df, columns=['Make', 'Model', 'Vehicle Class','Fuel Type'])
```

```python
df = df.astype(int, errors='ignore')
```

```python
df
```

```python
x = df.drop(columns='CO2 Emissions(g/km)')
y = df['CO2 Emissions(g/km)']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
si = StandardScaler()
x_train_scaled = si.fit_transform(x_train)
x_test_scaled = si.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
ri = LinearRegression()
ri.fit(x_train_scaled, y_train)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
y_pred = ri.predict(x_test_scaled)
r2 = r2_score(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print(f"RÂ² Score: {r2:.4f}")
print(f"RMSE: {rmse:.2f} g/km")
```
